```python
# 載入需要的套件
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
plt.style.use('ggplot')

# 忽略警告訊息
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
```

1. 請用 numpy 建立一個 10 x 10, 數值分布自 -1.0 ~ 1.0 的矩陣並繪製 Heatmap

```python
matrix = (np.random.random((10,10)) - 0.5) * 2  #原本隨機 0.0 ~ 1.0, 減 0.5 乘兩倍 -> -1.0 ~ 1.0

plt.figure(figsize=(10,10))
heatmap = sns.heatmap(matrix , cmap = plt.cm.RdYlBu_r, vmin = -1.0, annot = True, vmax = 1)
plt.title('Correlation Heatmap');
plt.show()
```

2. 請用 numpy 建立一個 1000 x 3, 數值分布為 -1.0 ~ 1.0 的矩陣，並繪製 PairPlot (上半部為 scatter, 對角線為 hist, 下半部為 density)

```python
nrow = 1000
ncol = 3

matrix = (np.random.random((nrow ,ncol)) - 0.5)*2

indice = np.random.choice([0,1,2], size=nrow) #隨機給予 0, 1, 2 三種標籤
plot_data = pd.DataFrame(matrix, indice)

# 繪製 seborn 進階 Heatmap
grid = sns.PairGrid(data = plot_data, size = 3, diag_sharey=False)
grid.map_upper(plt.scatter , alpha = 0.2)  #上半部為 scatter
grid.map_diag(sns.kdeplot) #對角線畫 histogram
grid.map_lower(sns.kdeplot , cmap = plt.cm.OrRd_r) #下半部放 density plot

plt.show()
```

3. 請用 numpy 建立一個 1000 x 3, 數值分布為常態分佈的矩陣，並繪製 PairPlot (上半部為 scatter, 對角線為 hist, 下半部為 density)

```python
nrow = 1000
ncol = 3

matrix = np.random.randn(nrow * ncol).reshape((nrow, ncol))

# 隨機給予 0, 1, 2 三種標籤
indice = np.random.choice([0,1,2], size=nrow)
plot_data = pd.DataFrame(matrix, indice)

# 繪製 seborn 進階 Heatmap
grid = sns.PairGrid(data = plot_data, size = 3, diag_sharey=False)
grid.map_upper(plt.scatter , alpha = 0.2)  #上半部為 scatter
grid.map_diag(sns.kdeplot) #對角線畫 histogram
grid.map_lower(sns.kdeplot , cmap = plt.cm.OrRd_r) #下半部放 density plot

plt.show()
```
